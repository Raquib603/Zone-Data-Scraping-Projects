```python
from bs4 import BeautifulSoup
import csv

# Read the saved HTML
with open("Zillow_Rental_Property.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

# Prepare lists
listing_urls = []
property_addresses = []
property_prices = []

# Extract each property card
cards = soup.find_all("article")  # Zillow wraps each listing in <article>
for card in cards:
    # URL
    url_tag = card.find("a", href=True)
    url = url_tag['href'] if url_tag else "N/A"
    if url != "N/A" and not url.startswith("http"):
        url = "https://www.zillow.com" + url
    listing_urls.append(url)

    # Address
    address_tag = card.find("address")
    address = address_tag.get_text(strip=True) if address_tag else "N/A"
    property_addresses.append(address)

    # Price
    price_tag = card.find("span", {"data-test": "property-card-price"})
    price = price_tag.get_text(strip=True) if price_tag else "N/A"
    property_prices.append(price)

# Save to CSV
with open("Zillow_Rentals.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["URL", "Address", "Price"])
    for url, addr, price in zip(listing_urls, property_addresses, property_prices):
        writer.writerow([url, addr, price])

print("Zillow_Rentals.csv created successfully!")
```

**Reasoning**:
The previous step successfully extracted listing URLs, property addresses, and prices into separate lists. To complete the subtask of storing this data in a pandas DataFrame, I will now create a DataFrame from these lists.



```python
df = pd.DataFrame({
    'Listing URL': listing_urls,
    'Property Address': property_addresses,
    'Property Price': property_prices
})

print("DataFrame created successfully:")
print(df.head())
```

```python
import json
import csv
from bs4 import BeautifulSoup

# Load the saved HTML
with open("Zillow_Rental_Property.html", "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

# Find the JSON data
script_tag = soup.find('script', id='__NEXT_DATA__')
if not script_tag:
    print("No __NEXT_DATA__ script found")
    exit()

json_data = json.loads(script_tag.string)

# Try common paths for listings
results = json_data.get('props', {}).get('pageProps', {}).get('searchPageState', {}).get('cat1', {}).get('searchResults', {}).get('listResults', [])

print(f"Found {len(results)} listings.")

# Extract required fields
data_list = []
for listing in results:
    # URL
    url = listing.get('detailUrl', 'N/A')
    if url and not url.startswith('http'):
        url = 'https://www.zillow.com' + url

    # Address
    address = listing.get('address', 'N/A')
    if isinstance(address, dict):
        address = address.get('streetAddress', 'N/A')

    # Price — first unit only
    price = 'N/A'
    units = listing.get('units', [])
    if units and isinstance(units, list):
        price = units[0].get('price', 'N/A')

    # Bedrooms — from first unit
    beds = 'N/A'
    if units and isinstance(units, list):
        beds = units[0].get('beds', 'N/A')

    # Bathrooms — not in `units` in your snippet, so keep N/A
    baths = 'N/A'

    data_list.append([url, address, price, beds, baths])

# Save to CSV
with open("Zillow_Rentals.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["URL", "Address", "Price", "Beds", "Baths"])
    writer.writerows(data_list)

print(f"Saved {len(data_list)} listings to Zillow_Rentals.csv")
```
