####Get the images

```python
links = []

for a in soup.find_all("a"):
    img = a.find("img")
    if img and img.get("src"):
        links.append(img["src"])

print(links)
```

```python
soup.find_all("a")
```

```python
for i in soup.find_all("a"):
  print(i)
```

```python
for i in soup.find_all("a"):
  print(i.get("href"))
```

#Save links in CSV

```python
#Collect links into a list

links = []

for i in soup.find_all("a"):
    href = i.get("href")
    if href:
        links.append(href)
```

```python
#write csv
import csv

with open("links.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["url"])  # header
    for link in links:
        writer.writerow([link])
```

```python
#download csv
from google.colab import files


with open("links.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["url"])
    for link in links:
        writer.writerow([link])

files.download("links.csv")
```
